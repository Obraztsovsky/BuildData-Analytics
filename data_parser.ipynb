```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
import random
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

intangible_assets_old=[]
intangible_assets_new=[]
fixed_assets_new=[]
fixed_assets_old=[]
prof_invest_new=[]
prof_invest_old=[]
profit_new=[]
profit_old=[]

f = open('new_table.txt', 'w')

with open('links.txt', 'r', encoding='utf-8') as a:
    links = a.read().splitlines()

link=links[1]
driver = webdriver.Edge()
driver.maximize_window()
driver.get(link)
time.sleep(4)
window_size = driver.get_window_size()
width = window_size['width']
height = window_size['height']
random_x = random.randint(100, width - 100)
random_y = random.randint(100, height - 100)    
actions = ActionChains(driver)
actions.move_by_offset(random_x, random_y).click().perform()
time.sleep(0.5)
actions.move_by_offset(-random_x, -random_y).perform()

for link in links:
    driver.get(link)
    try:
        element = WebDriverWait(driver, 15).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="balance"]/button'))
        )
        
        WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="balance"]/button'))
        )
        element.click()
    except:
        continue
    time.sleep(0.1)
    try:
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[2]/div[2]/div/div/div[2]/div[2]/div[2]/div/div[1]')
        element.click()
        time.sleep(0.1)
    except:
        print("")
        
    try:
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[2]/div[2]/div/div/div[2]/div[2]/div[2]/div/div[2]/div[3]')
        text=element.text
        number = text.split('\n')[0]
        intangible_assets_new.append(number)
        f.write(number+';')
    except:
        intangible_assets_new.append(0)
        f.write('0'+';')

    try:
        time.sleep(0.1)
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[2]/div[2]/div/div/div[2]/div[2]/div[2]/div/div[2]/div[4]')
        text=element.text
        number = text.split('\n')[0]
        intangible_assets_old.append(number)
        f.write(number+';')
    except:
        intangible_assets_old.append(0)
        f.write('0'+';')

    try:
        time.sleep(0.1)
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[2]/div[2]/div/div/div[2]/div[2]/div[2]/div/div[6]/div[3]')
        text=element.text
        number = text.split('\n')[0]
        fixed_assets_new.append(number)
        f.write(number+';')
    except:
        fixed_assets_new.append(0)
        f.write('0'+';')

    try:
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[2]/div[2]/div/div/div[2]/div[2]/div[2]/div/div[6]/div[4]')
        text=element.text
        number = text.split('\n')[0]
        fixed_assets_old.append(number)
        f.write(number+';')
    except:
        fixed_assets_old.append(0)
        f.write('0'+';')
        
    try:
        time.sleep(0.1)
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[2]/div[2]/div/div/div[2]/div[2]/div[2]/div/div[7]/div[3]')
        text=element.text
        number = text.split('\n')[0]
        prof_invest_new.append(number)
        f.write(number+';')
    except:
        prof_invest_new.append(0)
        f.write('0'+';')

    try:
        time.sleep(0.1)
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[2]/div[2]/div/div/div[2]/div[2]/div[2]/div/div[7]/div[4]')
        text=element.text
        number = text.split('\n')[0]
        prof_invest_old.append(number)
        f.write(number+';')
    except:
        prof_invest_old.append(0)
        f.write('0'+';')
        
    try:
        element = driver.find_element(By.XPATH, '//*[@id="financialResult"]/button')
        element.click()
        time.sleep(0.2)
    except:
        print("")
    try:
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[3]/div[2]/div/div/div[2]/div[2]/div/div[6]/div[3]')
        text=element.text
        number = text.split('\n')[0]
        profit_new.append(number)
        f.write(number+';')
    except:
        profit_new.append(0)
        f.write('0'+';')

    try:
        element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div[2]/div[2]/div/div[2]/div[3]/div[2]/div/div/div[2]/div[2]/div/div[1]/div[4]')
        text=element.text
        number = text.split('\n')[0]
        profit_old.append(number)
        f.write(number+'\n')
    except:
        profit_old.append(0)
        f.write('0'+'\n')

f.close()
a.close()
```
