```python
df1=pd.read_csv("C:/Users/gosha/Анализ строительной отрасли/new_table.csv",sep=';',names=['Нематериальные активы 2024','Нематериальные активы 2023','Основные средства 2024','Основные средства 2023','Доходные вложения 2024','Доходные вложения 2023','Доход 2024','Доход 2023'])
df1.head()
```

```python
import pandas as pd

def clean_data(value):
    
    if pd.isna(value):
        return 0
    
    if value == '-' or value == 0 or value == '0':
        return 0
    
    if isinstance(value, (int, float)):
        return int(value)
    
    if isinstance(value, str):
        value = value.replace(' ', '')
        
        if value.startswith('(') and value.endswith(')'):
            inner_value = value[1:-1]
            try:
                return -int(inner_value)
            except ValueError:
                inner_value = inner_value.replace(' ', '')
                return -int(inner_value)
        try:
            return int(value)
        except ValueError:
            return 0
    
    return 0

def clean_df(df):
    return df.map(clean_data)

df1=clean_df(df1)
df1.head()
```

```python
df_final = pd.concat([df, df1], axis=1)
df_final.tail()
```

```python
df_final=df_final[df_final['Доход 2024']!=0]
df_final=df_final[df_final['Доход 2023']!=0]
```

```python
df_final['Амортизация НМА(приблизительно)'] = df_final['Нематериальные активы 2024']-df_final['Нематериальные активы 2023']
df_final['Амортизация вся(приблизительно)'] = df_final['Нематериальные активы 2024']-df_final['Нематериальные активы 2023'] +\
df_final['Основные средства 2024'] - df_final['Основные средства 2023'] + df_final['Доходные вложения 2024'] - df_final['Доходные вложения 2023']
df_final['ebita']=df_final['Доход 2024'] + df_final['Амортизация НМА(приблизительно)'] 
df_final['ebitda']=df_final['Доход 2024'] + df_final['Амортизация вся(приблизительно)'] 
df_final.loc[df_final['Амортизация НМА(приблизительно)'] < 5, 'Амортизация НМА(приблизительно)'] = 0
df_final.loc[df_final['Амортизация вся(приблизительно)'] < 5, 'Амортизация вся(приблизительно)'] = 0
```
