```python
inns=[]
for x in df['ИНН']:
    inns.append(int(x))
```

```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.by import By
import random
import time
links=[]
f = open('links.txt', 'w')
driver = webdriver.Edge()
inn = "5038038838"
link = f"https://bo.nalog.gov.ru/search?query={inn}"
driver.get(link)
time.sleep(3)
window_size = driver.get_window_size()
width = window_size['width']
height = window_size['height']
random_x = random.randint(100, width - 100)
random_y = random.randint(100, height - 100)    
actions = ActionChains(driver)
actions.move_by_offset(random_x, random_y).click().perform()
time.sleep(0.5)
actions.move_by_offset(-random_x, -random_y).perform()
for inn in inns:
    link = f"https://bo.nalog.gov.ru/search?query={inn}"
    driver.get(link)
    time.sleep(0.2)
    try:
        link_element = driver.find_element(By.XPATH, '//*[@id="root"]/main/div/div/div[2]/div[2]/a')
        href = link_element.get_attribute("href")
        #links.append(href)
        f.write(href+'\n')
    except Exception:
        print(f"Информации для инн {inn} нет")
        df=df[df['ИНН']!=inn]
f.close()
```
